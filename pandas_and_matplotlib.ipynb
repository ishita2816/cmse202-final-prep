```python
import matplotlib.pyplot as plt
#basic plot
years = [1950, 1970, 1990, 2010]
population = [2.5, 3.7, 5.3, 7.0]

plt.plot(years, population)  # defines the line chart
plt.show()  # displays the plot
```

```python
#scatter plot
plt.scatter([1950, 1970, 1990, 2010], [2.5, 3.7, 5.3, 7.0])
plt.xlabel("Year")
plt.ylabel("World Population (billions)")
plt.title("World Population Scatter Plot")
plt.show()
```

```python
#log scale 


x = [1, 10, 100, 1000, 10000]
y = [10, 100, 1000, 10000, 100000]

plt.plot(x, y, marker='o')
plt.yscale('log')  # set y-axis to logarithmic scale
plt.xlabel("X values")
plt.ylabel("Y values (log scale)")
plt.title("Line Plot with Logarithmic Y-axis")
plt.show()

```

```python
values = [0.5, 1, 1.5, 2, 2.5, 2.8, 3, 3.2, 3.5, 4, 5, 5.5]

plt.hist(values, bins=3)
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
year = [1950, 1970, 1990, 2010, 2030, 2050, 2100]
pop = [2.5, 3.7, 5.3, 7.0, 8.5, 9.7, 10.0]
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("World Population (billions)")
plt.title("World Population Projections")
plt.show()
```

```python
# Add more points to give context or show trends over longer periods.
year = [1800, 1850, 1900] + year
pop = [1.0, 1.2, 1.6] + pop

plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("World Population (billions)")
plt.title("World Population Projections with Historical Data")
plt.yticks([0, 2, 4, 6, 8, 10], ["0", "2B", "4B", "6B", "8B", "10B"])     #positions and labels 
plt.show()
```

```python
# Define dictionary
brics = {
    "labels": ["BZ", "RU", "IN", "CH", "SA"],
    "country": ["Brazil", "Russia", "India", "China", "South Africa"],
    "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],        # example values
    "population": [200.4, 143.5, 1252, 1357, 52.98]    # example values
}

# Convert to DataFrame
brics_df = pd.DataFrame(brics)

print(brics_df)
```

```python
#print a column 
brics_df["country"]
```

```python
#print column as a dataframe 
brics_df[["country"]]
```

```python
#multiple columns 
brics_df[["country", "capital"]]
```

```python
#accessing rows using slices 
brics_df[1:4]
```

```python
#loc- label based access 
brics_df.index = ["BR", "RU", "IN", "CH", "SA"]
brics_df.loc["RU"]
brics_df.loc[["RU", "IN", "CH"]]
#all rows +selected columns 
brics_df.loc[:, ["country", "capital"]]
```

```python
#iloc- position based 
brics_df.iloc[1]
#multiple rows by index
brics_df.iloc[1:4]
```

```python
# all rows + selected columns 
brics_df.iloc[:, [0, 1]]
```
