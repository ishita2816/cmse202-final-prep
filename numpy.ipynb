```python
sudoku_list=[[0, 0, 4, 3, 0, 0, 2, 0, 9],
[0, 0, 5, 0, 0, 9, 0, 0, 1],
 [0, 7, 0, 0, 6, 0, 0, 4, 3],
 [0, 0, 6, 0, 0, 2, 0, 8, 7],
 [1, 9, 0, 0, 0, 7, 4, 0, 0],
 [0, 5, 0, 0, 8, 3, 0, 0, 0],
 [6, 0, 0, 0, 0, 0, 1, 0, 5],
 [0, 0, 3, 5, 0, 8, 6, 9, 0],
 [0, 4, 2, 9, 1, 0, 3, 0, 0]]
sudoku_array = np.array(sudoku_list)
sudoku_solution_list = [
    [8, 1, 4, 3, 7, 6, 2, 5, 9],
    [3, 2, 5, 4, 0, 9, 8, 7, 1],
    [9, 7, 1, 8, 6, 5, 0, 4, 3],
    [4, 3, 6, 1, 0, 2, 9, 8, 7],
    [1, 9, 8, 6, 0, 7, 4, 3, 2],
    [2, 5, 7, 0, 8, 3, 6, 1, 4],
    [6, 8, 9, 7, 3, 4, 1, 2, 5],
    [7, 1, 3, 5, 2, 8, 6, 9, 4],
    [5, 4, 2, 9, 1, 6, 3, 7, 8]
]
sudoku_solution= np.array((sudoku_solution_list))
```

```python
breaches = np.array([
    [2, 0, 3],   # year 1
    [1, 5, 0],   # year 2
    [0, 2, 1],   # year 3
    [3, 1, 1],   # year 4
    [0, 1, 1]    # year 5
])
breaches.sum(axis=0)   # column totals
breaches.sum(axis=1)   # row totals
```

```python
breaches.sum()
```

```python
#minimum and maximum
breaches.min()   # 0
breaches.max()   # 5
#axis=0 for columns here!!!!
breaches.min(axis=0)  # array([0, 0, 0])
breaches.max(axis=0)  # array([3, 5, 3])
```

```python
#row wise axis=1
breaches.min(axis=1)
# array([0, 0, 0, 1, 0])
```

```python
#over all mean 
breaches.mean()
```

```python
#mean across rows
breaches.mean(axis=1)
```

```python
#keepdims= Keeps reduced dimensions as size 1, useful for concatenation or broadcasting.
breaches.sum(axis=1, keepdims=True)
```

```python
#cummulative sum across columns
breaches.cumsum(axis=0)
```

```python
#example 
monthly_sales= np.array([[ 4134, 23925,  8657],
       [ 4116, 23875,  9142],
       [ 4673, 27197, 10645],
       [ 4580, 25637, 10456],
       [ 5109, 27995, 11299],
       [ 5011, 27419, 10625],
       [ 5245, 27305, 10630],
       [ 5270, 27760, 11550],
       [ 4680, 24988,  9762],
       [ 4913, 25802, 10456],
       [ 5312, 25405, 13401],
       [ 6630, 27797, 18403]])
# Create a 2D array of total monthly sales across industries
monthly_industry_sales = monthly_sales.sum(axis=1, keepdims=True)
print(monthly_industry_sales)

# Add this column as the last column in monthly_sales
monthly_sales_with_total = np.concatenate((monthly_sales, monthly_industry_sales), axis=1)
print(monthly_sales_with_total)
```

```python
# Create the 1D array avg_monthly_sales
avg_monthly_sales = monthly_sales.mean(axis=1)
print(avg_monthly_sales)

# Plot avg_monthly_sales by month
plt.plot(np.arange(1, 13), avg_monthly_sales, label="Average sales across industries")

# Plot department store sales by month
plt.plot(np.arange(1, 13), monthly_sales[:, 2], label="Department store sales")
plt.legend()
plt.show()
```

```python
#vectorize function- Wraps Python function  makes it work elementwise 
arr = np.array(["hi", "apple", "yo"])
len(arr)  
# returns 3- Not vectorized (wrong)
```

```python
vlen = np.vectorize(len)
vlen(arr)
# array([2, 5, 2])- vectorized
```

```python
names = np.array(["alice", "bob", "charlie"])
# Vectorize the .upper() string method
vectorized_upper = np.vectorize(str.upper)

# Apply vectorized_upper to the names array
uppercase_names = vectorized_upper(names)
print(uppercase_names)
```

```python
rgb_array = np.random.randint(0, 256, (4, 4, 3))
# Split the RGB array along the color axis
red, green, blue = np.split(rgb_array, 3, axis=2)

print("Red array shape:", red.shape)  # (4, 4, 1)
```

```python
#np.stack() combines arrays along a new axis
# Create three 2D arrays (like color channels)
red = np.array([[255, 0],
                [0, 255]])

green = np.array([[0, 255],
                  [255, 0]])

blue = np.array([[0, 0],
                 [255, 255]])

# Stack them along a new 3rd axis to make a 3D RGB array
rgb_array = np.stack([red, green, blue], axis=2)

print("Stacked RGB array:\n", rgb_array)
print("Shape:", rgb_array.shape)  # Shape is (2, 2, 3)
```
