```python
#employee class using init 
class Employee:
    """A class to represent an employee."""

    def __init__(self, name, salary):
        self.name = name      # attribute set once during object creation
        self.salary = salary  # attribute set once during object creation

    def give_raise(self, amount):
        """Increase the salary by amount."""
        self.salary += amount

    def monthly_salary(self):
        """Return the monthly salary (1/12th of annual)."""
        return self.salary / 12


# Create an employee object
emp = Employee("Korel Rossi", 50000)

# Access attributes
print(emp.name)     # Korel Rossi
print(emp.salary)   # 50000

# Give a raise
emp.give_raise(1500)
print(emp.salary)   # 51500

# Get monthly salary
print(emp.monthly_salary())  # 4291.666...
```
